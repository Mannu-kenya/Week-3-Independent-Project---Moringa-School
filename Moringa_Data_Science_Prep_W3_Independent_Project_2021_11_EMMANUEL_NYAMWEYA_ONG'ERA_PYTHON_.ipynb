##Loading Telcom_dataset2

```python
#loading the Telcom_dataset2.csv file 

Telcom_dataset2 = pd.read_csv('/content/Telcom_dataset2.csv')

Telcom_dataset2 
```

##Loading our Telcom_dataset3##

```python
Telcom_dataset3 = pd.read_csv('/content/Telcom_dataset3.csv')

Telcom_dataset3
```

##Renaming the geo-cells table with the names given cells_geo_description.xlsx##

```python
cells_geo_1= cells_geo[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].apply(lambda x: pd.Series([i for i in x.lower().split(';'or';;')]))
# The split() method splits a string into a list.
cells_geo_1
```

##Renaming the colums as per the description##

```python
cells_geo_2= cells_geo_1.rename(columns = {1:'CITY',2:'STATUS', 3:'LOCALISATION', 4:'DECOUPZONE', 5:'ZONENAME', 6:'LONGITUDE', 7:'LATITUDE', 8:'REGION', 9:'AREA', 10:'CELL_ID', 11:'SITE_CODE'})

cells_geo_2
```

##Dropping the column zero which has been duplicated##

```python
cells_geo_3 = cells_geo_2.drop([0], axis = 1)

cells_geo_3
```

##Renaming datetime and the mispelt Product name ##

```python
Telcom_dataset1 = Telcom_dataset1.rename(columns = {'PRODUTC':'PRODUCT', 'DATETIME':'DATE_TIME'})
Telcom_dataset1

# We then rename DW_B_NUMBER & DW_A_NUMBER to DW_B_NUMBER_INT & DW_A_NUMBER_INT in the second da

Telcom_dataset2 = Telcom_dataset2.rename(columns = {'DW_B_NUMBER':'DW_B_NUMBER_INT', 'DW_A_NUMBER':'DW_A_NUMBER_INT' })
Telcom_dataset2

#We rename cell id and mispelt site id in data 3 

Telcom_dataset3= Telcom_dataset3.rename(columns = {'CELLID':'CELL_ID', 'SIET_ID':'SITE_ID' })
Telcom_dataset3

#Then we merge the two dataframes 

df_for_two_days = pd.concat((Telcom_dataset1, Telcom_dataset2), axis=0, ignore_index=True)

df_for_two_days

# Then we merge df_for_two_days with df_telcom_day_3

df_for_three_days = pd.concat((df_for_two_days, Telcom_dataset3), axis=0, ignore_index=True)

#Preview the combined table 

df_for_three_days
```

##Merging df_for_3_days with cell geo data##

```python
cells_geo_3=cells_geo_3.rename(columns = {'SITE_CODE':'SITE_ID'})

main_dataframe=pd.concat((df_for_three_days,cells_geo_3), axis=0, ignore_index=True)

main_dataframe

main_dataframe_1=main_dataframe.drop_duplicates()

main_dataframe_2=main_dataframe_1.fillna(0)

main_dataframe_2
```

##Which ones were the most used city for the three days?##
